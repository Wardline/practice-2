```python
class Chocolate:
    def __init__(self, price: float, addition: str = None):
        # Проверка корректности цены
        if price <= 0:
            raise ValueError("Цена шоколадного батончика должна быть больше нуля")
        self.price = price
        self.addition = addition

    def show_my_chocolate(self):
        if self.addition:
            print(f"Шоколадный батончик, добавка: {self.addition}")
        else:
            print("Обычный шоколадный батончик")

    def __str__(self):
        return f"Шоколадный батончик, {self.price} рублей"
```

```python
# 1. Обычный шоколадный батончик
choco1 = Chocolate(75.0)
choco1.show_my_chocolate()
print(choco1)
```

```python
# 2. Батончик с добавкой
choco2 = Chocolate(90.5, "орехи")
choco2.show_my_chocolate()
print(choco2)
```

```python
# 3. Батончик с другой добавкой
choco3 = Chocolate(110.0, "карамель")
choco3.show_my_chocolate()
print(choco3)
```

```python
# 4. Пример с ошибкой (для проверки)
try:
    bad_choco = Chocolate(-50)
except ValueError as e:
    print("Ошибка:", e)
```
